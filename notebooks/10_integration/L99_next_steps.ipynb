## 1. Aurora v1 的边界

Aurora v1 证明了：

| 能力 | 证据 |
|---|---|
| 从零实现核心 DSP 算法 | FFT / STFT / Mel / MFCC 全部有测试 |
| 理解深度学习基础 | autograd 从零，KWS 训练跑通 |
| 使用预训练模型（pre-trained model）并评估 | Whisper 微调（fine-tuning） + WER 数字 |
| 端到端系统集成 | 音频输入 → 转录 / 推荐 → 输出 |
| 技术沟通 | 论文阅读、技术报告、白板推导 |

Aurora v1 **没有**证明的（诚实）：

| 未覆盖 | 原因 |
|---|---|
| 实时流式处理（< 200ms 延迟） | 需要 C++/CUDA kernel |
| 大规模训练（> 1B 参数） | 需要多节点 GPU 集群 |
| TTS / 声音克隆（voice cloning，TTS） | 需要 vocoder 专项研究 |
| Production MLOps | 需要 K8s / 监控系统实战 |

## 2c. 音频 AI 的开放研究问题

如果你选择方向 A 或想找论文切入点，以下是截至 2026 年仍有探索空间的方向：

| 问题 | 现状 | 可能的切入 |
|---|---|---|
| 低资源语言 ASR | 数据稀缺，WER 仍高 | 数据增强 + 跨语言迁移 |
| 音乐理解 | 和声、节奏分析准确率低 | 改进 chroma / beat 特征 |
| 实时 TTS 音色克隆 | 延迟 > 500ms，音质 vs 速度 tradeoff | 流式 vocoder |
| 音频事件检测 | 多事件重叠时准确率下降 | 多标签分类（multi-label classification） + 时间定位 |
| ASR 鲁棒性 | 噪声环境 WER 急剧上升 | 数据增强（SpecAugment 扩展） |
| 音频 Foundation Model 评估 | 缺乏统一 benchmark | 类 GLUE 的音频评测集 |

这些方向都有入门门槛不高的子问题——一个好的毕业论文 / 会议短文
往往只需要**在一个具体数据集上提升一个指标 1-3%**，并有清晰的 ablation。

## 3c. 如何建立公开形象（不需要成为网红）

公开形象不是营销，是让对的人找到你。

## 3g. 长期视角：2 年后的两条路

## 4. 持续成长的习惯系统

课程结束后，保持成长最有效的三个习惯：
