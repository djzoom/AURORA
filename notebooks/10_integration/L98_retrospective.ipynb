## 1. 交付物清单

按模块自评。评分标准：
- ✅ **Done**：有代码、有测试、能白板推导

- ❌ **Missing**：没做到

```python
deliverables = {
    # ─── Audio Core（Month 1 目标）───────────────────────────────────
    "FFT（从空白文件实现，测试全绿）":          None,  # ✅ / 🔄 / ❌
    "STFT（实现 + 窗函数 + overlap-add）":     None,
    "Mel filterbank（Hz→Mel 映射 + 矩阵）":    None,
    "MFCC（DCT-II + 系数切片）":               None,
    "真实音频输入（LibriSpeech WAV 跑通）":     None,

    # ─── Deep Learning（Month 2 目标）───────────────────────────────
    "从零 autograd（forward + backward）":     None,
    "KWS 分类器（Mel 特征 + 训练曲线）":       None,

    # ─── ASR（Month 3 目标）─────────────────────────────────────────
    "CTC 前向算法（手写）":                    None,
    "Whisper 微调 + WER 评估":                 None,

    # ─── Music（Month 4 目标）───────────────────────────────────────
    "chroma / onset / beat 特征":              None,
    "音乐 embedding + k-NN 推荐":              None,

    # ─── LLM / RAG（Month 5 目标）───────────────────────────────────
    "Transformer attention（从零）":           None,
    "KV-Cache 实现":                           None,
    "RAG 流水线（TF-IDF + 向量检索）":         None,

    # ─── Integration（Month 6 目标）─────────────────────────────────
    "端到端 demo（可演示）":                   None,
    "技术报告 / arXiv 预印本":                 None,
}

done    = sum(1 for v in deliverables.values() if v == "✅")
partial = sum(1 for v in deliverables.values() if v == "🔄")
missing = sum(1 for v in deliverables.values() if v == "❌")
total   = len(deliverables)

print(f'完成 {done}/{total}   部分 {partial}   缺失 {missing}')
print()
for k, v in deliverables.items():
    if v is not None:
        print(f'  {v}  {k}')
    else:
        print(f'  ?   {k}  ← 未填')
```

## 6. 学习速度复盘：预估 vs 实际

每个月的进度和预期相比如何？诚实记录有助于你下一个项目更准确地估时。

```python
learning_velocity = {
    # 格式：(预估天数, 实际天数, 最难的一个概念)
    "Month 1 · Audio Core":       (30, None, None),
    "Month 2 · Deep Learning":    (30, None, None),
    "Month 3 · ASR":              (30, None, None),
    "Month 4 · Music":            (30, None, None),
    "Month 5 · LLM/RAG":         (30, None, None),
    "Month 6 · Integration":      (30, None, None),
}

print("月份            预估   实际   最难概念")
print("-" * 60)
for phase, (est, actual, hard) in learning_velocity.items():
    actual_str = str(actual) if actual is not None else "?"
    hard_str   = hard if hard is not None else "未填"
    ratio = f"({actual/est*100:.0f}%)" if actual is not None else ""
    print(f"{phase[:22]:<22} {est:>4}天  {actual_str:>4}天 {ratio:>6}  {hard_str}")
```

```python
# 成长曲线可视化：预估 vs 实际天数对比（满足副标题「成长曲线量化」承诺）
# 保护：仅当所有 actual 字段已填时自动绘图
if all(v[1] is not None for v in learning_velocity.values()):
    try:
        import matplotlib.pyplot as plt
        import matplotlib
        matplotlib.rcParams['font.family'] = ['Arial Unicode MS', 'Hiragino Sans GB', 'DejaVu Sans']

        phases = list(learning_velocity.keys())
        labels = [p.split('·')[1].strip() if '·' in p else p for p in phases]
        estimated = [v[0] for v in learning_velocity.values()]
        actual    = [v[1] for v in learning_velocity.values()]

        x = list(range(len(phases)))
        fig, ax = plt.subplots(figsize=(9, 4))
        width = 0.35
        bars_est = ax.bar([xi - width/2 for xi in x], estimated, width,
                          label='预估天数', color='#4C72B0', alpha=0.85)
        bars_act = ax.bar([xi + width/2 for xi in x], actual,    width,
                          label='实际天数', color='#DD8452', alpha=0.85)

        ax.set_xticks(x)
        ax.set_xticklabels(labels, rotation=20, ha='right')
        ax.set_ylabel('天数')
        ax.set_title('Aurora 6 个月学习速度：预估 vs 实际（成长曲线量化）')
        ax.legend()
        ax.axhline(30, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)

        for bar in bars_est:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                    str(int(bar.get_height())), ha='center', va='bottom', fontsize=8)
        for bar in bars_act:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                    str(int(bar.get_height())), ha='center', va='bottom', fontsize=8)

        plt.tight_layout()
        plt.show()
    except ImportError:
        print("matplotlib 未安装，跳过可视化。运行 pip install matplotlib 后重试。")
else:
    print("📊 可视化待激活：请先在 learning_velocity 中填写所有 actual（实际天数）字段")
    for phase, (est, actual, hard) in learning_velocity.items():
        if actual is None:
            print(f"  ⬜ {phase}：实际天数未填")
```
