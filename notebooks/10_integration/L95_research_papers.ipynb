## 2. 如何高效阅读论文——三遍法

### 第一遍（5-10 分钟）：定位
1. **标题 + 摘要**：这篇在解决什么问题？提出了什么方法？
2. **图表**：只看图，不看正文。图 1 通常是方法示意图，最后一张图通常是结果对比。
3. **结论**：作者自己说"我做到了什么"。
4. 决策：值得精读吗？

### 第二遍（1-2 小时）：理解
- 读引言（背景和动机）
- 精读方法章节：每个公式要能手推，每个符号要能叫出名字
- 对照代码（如果开源）：论文公式 ↔ 代码函数名
- 跳过你暂时看不懂的证明，标记后回来

### 第三遍（半天-1 天）：复现
- 用纸推导核心公式
- 从空白文件写出核心算法（Aurora 风格：no wrapper）
- 跑一个 toy 实验验证输出和论文一致

## 3. 学术论文的结构

一篇标准 8 页会议论文的骨架：

```
Abstract（~150词）
  └─ 问题 / 方法 / 关键结果 / 结论（各一句话）

1. Introduction（~1页）
  └─ 动机 → 现有方法的局限 → 我们的贡献（our contributions）

2. Related Work（~0.5页）
  └─ 把你的方法和 3-5 篇相关论文对比，说清楚"不同在哪"

3. Method（核心，~2-3页）
  └─ 问题定义（符号表）→ 架构/算法 → 关键公式

4. Experiments（~2页）
  └─ 数据集 / 评估指标 → 主实验结果表 → Ablation study

5. Conclusion（~0.5页）
  └─ 重申贡献，点出局限和未来方向

References
  └─ 8-30篇，用 BibTeX 管理
```

**技术报告 vs 论文**：如果你的工作还没到发表标准，先写技术报告（Tech Report）
发到 arXiv，是非常好的积累——Whisper 本身发布时也只是 arXiv 预印本（preprint）。

## 4. 投稿去哪里

### 音频 / 语音 AI 主要会议

| 会议 | 全称 | 截止（通常） | 接收率 | 适合 Aurora 的方向 |
|---|---|---|---|---|
| **ICASSP** | IEEE Int. Conf. on Acoustics, Speech and Signal Processing | 每年 9-10 月 | ~50% | 信号处理、特征提取、ASR |
| **Interspeech** | Annual Conf. of ISCA | 每年 3-4 月 | ~50% | 语音识别、合成、说话人 |
| **ICLR** | Int. Conf. on Learning Representations | 每年 9-10 月 | ~30% | 音频 + 深度学习交叉 |
| **NeurIPS** | Neural Information Processing Systems | 每年 5-6 月 | ~25% | 生成模型、Foundation Model |
| **ICML** | Int. Conf. on Machine Learning | 每年 1-2 月 | ~28% | 算法理论 + 音频 |

### arXiv——预印本服务器
- 不经同行评审（peer review），上传即公开，通常 1 个工作日后可见
- 地址：`arxiv.org`，分类 `cs.SD`（Sound）、`cs.LG`、`eess.AS`
- 策略：先发 arXiv 占坑，同时投会议——不冲突

### 审稿时间线（以 ICASSP 为例）
```
9 月  → 提交截止
11 月 → 审稿结果
12 月 → 修改/回复
1 月  → 最终通知
5 月  → 会议召开（现场 or 线上 poster/oral）
```

## 5. 文献管理：不要用文件夹堆 PDF

论文读多了，最大的痛点是「我在哪里看过这个结论」。
从一开始就用工具管理，后面写论文时引用会省去大量时间。

### 推荐工具

| 工具 | 优点 | 适合场景 |
|---|---|---|
| **Zotero**（免费） | 浏览器插件一键存、自动抓元数据、支持 BibTeX 导出 | 主力文献库，写论文必备 |
| **Obsidian + PDF** | 笔记和 PDF 联动、双向链接、本地存储 | 深度阅读笔记 |
| **Semantic Scholar** | AI 辅助相关论文推荐，引用网络可视化 | 发现相关论文 |
| **arXiv Sanity** | 按热度/相关性过滤 arXiv 新论文 | 每日跟进领域进展 |
| **Connected Papers** | 可视化论文引用关系图 | 摸清一个领域的脉络 |

### Aurora 的文献管理建议

```
docs/
  paper_notes/          ← 每篇论文一个 .md 文件
    whisper_2022.md     ← 三遍法笔记 + 关键公式 + 代码对应
    ctc_graves_2006.md
  references.bib        ← 统一的 BibTeX 文件，写报告时 \cite{} 直接调用
```

## 6. 同行评审：论文发表的核心机制

### 审稿流程（以双盲评审（double-blind review）为例）

```
投稿
  ↓ 1-2周
Area Chair 分配给 3 名 Reviewer
  ↓ 4-6周
3份审稿意见（每份通常 1-3 页）
  ↓ 1-2周
作者撰写 Rebuttal（回复，通常 ≤1页）
  ↓ 2周
Reviewer 可修改分数 → AC 给出推荐
  ↓ 1周
Program Committee 最终决定
```

### 审稿意见的三种类型

| 类型 | 信号 | 处理方式 |
|---|---|---|
| 技术错误 | 公式推导错、实验设计有缺陷 | 必须修正，感谢指出 |
| 缺少实验 | "没有和 X 方法对比" | 如果能跑，跑完加进去；否则说明原因 |
| 主观意见 | "我不觉得这个问题重要" | 礼貌解释动机，举引用证明重要性 |

### Rebuttal 的写法原则

1. **先感谢，再回应**——即使对方说错了
2. **每条意见单独回应**，不要混在一起
3. **用数字支撑**——"我们补跑了实验，结果提升 0.8% WER"比"我们认为..."更有说服力
4. **态度平和**——审稿人是志愿者，不是敌人

## 8. 与学术机构合作的实际路径

### 路径 1：开源复现
- 复现某实验室的论文，把代码发布到 GitHub，写详细的 README
- 作者通常会注意到，好的复现能带来联系机会
- Aurora 本身就是一个高质量复现项目

### 路径 2：联系导师 / 研究员
写邮件时的关键原则：**具体 > 热情**

```
主题: Re: [论文名] — 代码复现 + 一个问题

您好，

我最近在复现您的 [论文名] 中的 [具体方法]，
遇到了 [具体问题：如 Section 3.2 公式 (5) 的边界条件]。
我的实现见：[GitHub 链接]

另外，如果您组里有面向社区的开放实验（如数据标注、benchmark 测试），
我很乐意参与贡献。

[名字]
```

具体的问题 + 已有工作 → 回复率远高于"我想加入您的实验室"。

### 路径 3：参与开放数据集 / benchmark 项目
- **LibriSpeech**（OpenSLR）：开放 ASR 数据集，可提交 WER 结果
- **SUPERB**：语音任务 benchmark，提交系统跑分
- **HEAR**（Holistic Evaluation of Audio Representations）：音频 embedding 评测

### 路径 4：申请访问学生 / 远程合作
- 部分高校有"访问学生"身份，可跨机构合署论文
- 条件：有一定代码能力 + 有具体想法 + 导师有空间

## 9. 如何写 Related Work

Related Work 是最容易写坏的章节——很多人写成了"方法列表"。
好的 Related Work 要说清楚**你的工作和它们的关系**，不只是它们做了什么。

### 三个维度组织 Related Work

1. **按技术路线分组**（不按时间顺序）
   把方法分成 2-3 个流派，每组说共同点和局限，再说你怎么解决局限

2. **每篇文献一句差异**
   "Wang et al. (2022) 用 X 方法解决了 Y 问题，但在 Z 场景下精度下降 N%。
   本文通过 Q 改进了这一局限。"

3. **结尾一句定位**
   "与上述方法不同，本文是首个同时满足 A 和 B 约束的工作。"

### Aurora 场景的 Related Work 要覆盖什么

如果你写一篇关于 Aurora 音频特征提取的报告，Related Work 至少要提到：
- 传统 MFCC 方法（Davis & Mermelstein 1980）
- 端到端特征学习（wav2vec 2.0）
- 你的方法的定位：可解释、可推导、no-dependency

## 11. ✏️ 练习：标注一篇论文的结构

打开 Whisper 原文（或任意你手边的论文），
用下面的表格对照论文的实际章节，找出每个结构元素在哪一页。

````python
# ✏️ 填写论文结构定位（以你正在读的论文为准）
paper_structure_map = {
    "论文标题":         None,   # 填论文名
    "Abstract 页码":   None,   # 例: 1
    "问题定义所在位置": None,   # 例: "Section 2, p.2"
    "核心方法公式编号": None,   # 例: "Equation (3)"
    "主实验结果表格":   None,   # 例: "Table 2, p.6"
    "Ablation Study":  None,   # 例: "Table 4, p.7" 或 "无"
    "局限性说明":       None,   # 例: "Section 5, p.8" 或 "无"
    "Future Work":     None,   # 例: "Conclusion 末段"
}

filled = sum(1 for v in paper_structure_map.values() if v is not None)
print(f"结构定位：{filled}/{len(paper_structure_map)}")
for k, v in paper_structure_map.items():
    status = "✅" if v is not None else "⬜"
    print(f"  {status} {k}: {v or '（未填）'}")
````
