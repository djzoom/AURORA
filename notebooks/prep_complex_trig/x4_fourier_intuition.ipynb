## 1. 叠加几个正弦 → 复杂波形

```python
import numpy as np, matplotlib.pyplot as plt
t = np.linspace(0, 1, 500)
y = np.sin(2*np.pi*2*t) + 0.5*np.sin(2*np.pi*5*t)
plt.plot(t, y); plt.title('2 Hz + 0.5·(5 Hz)'); plt.show()
```
