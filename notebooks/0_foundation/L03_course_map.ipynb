## 1. Aurora 是什么

Aurora 是一个**从零开始**构建的音频 AI 研究系统。

它不是 API 拼装项目。这个区别很重要：

| 类型 | 例子 | 面试价值 |
|---|---|---|
| API 拼装 | Whisper API + OpenAI API + Next.js | 接近零——展示不了任何研究工程能力 |
| Aurora 的做法 | 自己写 FFT、自己训练模型、自己实现 Attention | 高——每一行都能在白板上推导 |

**核心原则**：当实现一个核心能力时，构建算法本身——不导入已经做好的黑盒。

具体来说：
- FFT、STFT、Mel 滤波器组、DCT、MFCC：全部手写，用 numpy.fft 当"对答案"工具验证
- 自动微分（autograd）：从零写，不 `import torch.autograd` 当捷径
- ASR：自己微调 Whisper，不调用 Whisper API
- LLM 推理：本地运行 Qwen/Llama，不调用 OpenAI API

**为什么这样做？** 音频/语音/音乐 AI 团队的面试官会问："你能推导 FFT 的复杂度吗？""反向传播的链式法则你能手写吗？""mel filterbank 为什么要三角形？" 只有真正实现过，才能回答这些问题。
