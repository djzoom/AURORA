## 7. 数组形状：shape 是代码的尺寸表 (shape)

`shape` 告诉我们数组有几个维度，每个维度有多长。

| 对象 | 例子 | shape |
|---|---|---|
| 标量 (scalar) | `3.0` | 没有 shape |
| 向量 (vector) | `[1, 2, 3]` | `(3,)` |
| 矩阵 (matrix) | `[[1, 2], [3, 4]]` | `(2, 2)` |


```python
v = np.array([1.0, 2.0, 3.0])
M = np.array([[1.0, 2.0],
              [3.0, 4.0]])

print('v.shape =', v.shape)
print('M.shape =', M.shape)
print('M @ np.array([10, 20]) =', M @ np.array([10.0, 20.0]))
```

## 10. 实践：改一个参数，看结果如何变化 (parameter experiment)

只改 `frequency`，观察同样 1 秒内波峰数量如何变化。


```python
sample_rate = 64
duration = 1.0
frequencies = [1.0, 3.0, 7.0]
t = make_time_axis(duration, sample_rate)

plt.figure(figsize=(8, 4))
for frequency in frequencies:
    x = make_sine(duration, sample_rate, frequency=frequency)
    plt.plot(t, x, label=f'{frequency} Hz')
plt.title('same duration, different frequencies')
plt.xlabel('time (s)')
plt.ylabel('amplitude')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

| duration | sample_rate | N | last time |
|---:|---:|---:|---:|
| 0.25 | 8 | ? | ? |
| 0.50 | 16 | ? | ? |
| 1.00 | 10 | ? | ? |

公式：

- `N = round(duration * sample_rate)`
- `last time = (N - 1) / sample_rate`


```python
cases = [
    (0.25, 8),
    (0.50, 16),
    (1.00, 10),
]

for duration, sample_rate in cases:
    t = make_time_axis(duration, sample_rate)
    print({
        'duration': duration,
        'sample_rate': sample_rate,
        'N': len(t),
        'last_time': t[-1],
    })
```
