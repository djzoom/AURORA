# 前导·微积分 3 — 链式法则（反向传播的心脏）

**目标**：复合函数 `f(g(x))` 的导数 = `f'(g(x)) · g'(x)`——一层层连乘。

**为什么对 Aurora 重要**：神经网络是"函数套函数套函数"。**反向传播 = 从输出往输入，把每层的导数连乘起来**。链式法则就是它的全部秘密。

## 1. 例子：y = sin(x²)

外层 f(u)=sin(u)，内层 g(x)=x²。
`dy/dx = cos(x²) · 2x`

```python
import numpy as np
x = 1.3
analytic = np.cos(x**2) * 2*x          # 链式法则
numeric = (np.sin((x+1e-5)**2) - np.sin((x-1e-5)**2)) / (2e-5)
print('链式法则:', round(analytic,5), ' 数值验证:', round(numeric,5))
```
