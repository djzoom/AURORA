## 1. 三条判据

1. **行列式 ≠ 0**（充要）
2. **所有特征值 ≠ 0**（充要）
3. **严格对角占优 (s.d.d.)**：每行对角元的绝对值 > 该行其余元素绝对值之和（**充分**但非必要）。

```python
import numpy as np
A = np.array([[1,1,0],[1,2,1],[0,1,3]], float)
print('det =', round(np.linalg.det(A)))
print('特征值:', np.round(np.linalg.eigvals(A), 3))
print('两条充要判据都说：可逆 =', abs(np.linalg.det(A))>1e-9)
```
