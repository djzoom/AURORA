## 1. L2 范数 = 向量的长度（勾股定理推广）

`|v| = sqrt(v1² + v2² + ...)`

```python
import numpy as np
v = np.array([3.0, 4.0])
print('手算:', np.sqrt(np.sum(v**2)))   # 5.0
print('numpy:', np.linalg.norm(v))      # 5.0
print('L1 范数(绝对值之和):', np.sum(np.abs(v)))
```

## 2. 两个向量的距离 = 差向量的长度

```python
a = np.array([1.0, 2.0]); b = np.array([4.0, 6.0])
print('距离 =', np.linalg.norm(a - b))  # 5.0
```
